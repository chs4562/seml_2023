```python
import pandas as pd # I m p o r t i e r e n der B i b l i o t h e k
import matplotlib.pyplot as plt
from plot_2d_classifier_functions import plot_2d_seperator
from sklearn.model_selection import train_test_split
data = pd.read_csv("database.csv") # E i n l e s e n d e s F i l e s
x = data.values[:,0:2] # Merkmalswerte
y = data.values[:,2] # K l a s s e n w e r t e
```

```python
data_x_1 = [0]*len(x)
data_x_2 = [0]*len(x)
data_y = [0]*len(x)
for n, dataset in enumerate(x):
    # print(n)
    data_x_1[n] = dataset[0]
    data_x_2[n] = dataset[1]
    data_y[n] = y[n]
```

```python
plt.scatter(data_x_1, data_x_2,c = data_y)
```

```python
test_data_size = 0.2
# sep = round(len(x)*(1-test_data_size))

# x_1_train = data_x_1[0:sep]
# x_1_test = data_x_1[sep:]

# x_2_train = data_x_2[0:sep]
# x_2_test = data_x_2[sep:]

# x_train = x[0:sep]
# x_test = x[sep:]

# y_train = data_y[0:sep]
# y_test = data_y[sep:]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y)

print ("Training dataset size: " + str(len(x_train)))
print ("Test dataset size: " + str(len(x_test)))
```

# Nearest Centroid Classifier

```python
from sklearn.neighbors import NearestCentroid
clf = NearestCentroid(metric='cosine')
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

# Naiver Bayes Klassifikator

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

# Entscheidungsbaum
### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'log_loss')
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

### Extra Tree Classifier

```python
from sklearn.tree import ExtraTreeClassifier
clf = ExtraTreeClassifier(criterion = 'gini', splitter = 'best')
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

# Random Forrest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 21, algorithm = 'auto', leaf_size = 20, metric = 'manhattan')
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```

# SVM

```python
from sklearn.svm import SVC
clf = SVC(kernel='poly', degree = 3, gamma = 100.0, coef0 = 1.0)
clf.fit(x_train, y_train)
print ("Score: " + str(clf.score(x_test, y_test)))
plot_2d_seperator(clf, x, y)
```
