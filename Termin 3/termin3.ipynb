```python
import pandas as pd 
dataframe = pd . read_csv ("database_training.csv")
```

```python
features = dataframe.loc[:, "v_x":"age"]
features
```

```python
X_onehot = pd.get_dummies(features)
X_onehot
```

```python
X = X_onehot.values
y = dataframe.label.values
print(X)
print(y)
```

```python
from matplotlib import pyplot as plt

df = pd.DataFrame(pd.np.random.randn(1000, 4), columns=['A long column name','B is also long','C is even longer ','D is short'])
axes = pd.plotting.scatter_matrix(X_onehot, c = y)
for ax in axes.flatten():
    ax.xaxis.label.set_rotation(90)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')

plt.tight_layout()
plt.gcf().subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler

X_best = SelectKBest(chi2, k=10).fit(X, y)
X_best.scores_
X_new = X_best.transform(X)

scaler = MinMaxScaler()
scaler.fit(X_new)
X_scaled = scaler.transform(X_new)
```

```python
from sklearn.decomposition import PCA
select = PCA(n_components=13)
select.fit(X)
X_new = select.transform(X)

scaler = MinMaxScaler()
scaler.fit(X_new)
X_scaled = scaler.transform(X_new)
```

```python
import numpy as np
plt.plot(np.cumsum(select.explained_variance_ratio_))
```

```python
print(X_scaled.shape)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from plot_2d_classifier_functions import plot_2d_seperator

cla = KNeighborsClassifier(n_neighbors=20, weights='distance')
cla.fit(X_train, y_train)
print("test score: ", cla.score(X_test, y_test))
print("train score: ", cla.score(X_train, y_train))
#plot_2d_seperator(cla, X_new, y)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = cla.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
cla = RandomForestClassifier(n_estimators=100, criterion='gini')
cla.fit(X_train, y_train)
print("test score: ", cla.score(X_test, y_test))
print("train score: ", cla.score(X_train, y_train))

y_pred = cla.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

parameters = {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['linear', 'poly', 'rbf', 'sigmoid'], 'coef0': [0.01, 0.1, 1, 10]}

svc = SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(X_train, y_train)
print("test score: ", clf.score(X_test, y_test))
print("train score: ", clf.score(X_train, y_train))

y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(clf.get_params())
print(sorted(clf.cv_results_.keys()))
```
